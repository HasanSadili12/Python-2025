```python
import time

# Status awal (bisa diganti untuk uji coba)
sensor_tangki_atas = False   # False = belum penuh, True = penuh
sensor_tangki_bawah = True   # False = kosong, True = ada air

# Status pompa
pompa_menyala = False

def nyalakan_pompa():
    global pompa_menyala
    if not pompa_menyala:
        pompa_menyala = True
        print("Pompa DINYALAKAN")
    else:
        print("Pompa sudah menyala")

def matikan_pompa():
    global pompa_menyala
    if pompa_menyala:
        pompa_menyala = False
        print("Pompa DIMATIKAN")
    else:
        print("Pompa sudah mati")

# Simulasi berjalan terus (bisa disetop dengan Ctrl+C)
try:
    while True:
        print("\n--- Mengecek kondisi ---")
        print(f"Tangki atas penuh?      {'Ya' if sensor_tangki_atas else 'Tidak'}")
        print(f"Tangki bawah ada air?   {'Ya' if sensor_tangki_bawah else 'Tidak'}")

        if sensor_tangki_bawah and not sensor_tangki_atas:
            nyalakan_pompa()
        else:
            matikan_pompa()

        # Simulasi perubahan kondisi (contoh)
        # Ganti kondisi untuk melihat perubahan
        time.sleep(3)

        # Contoh simulasi perubahan status sensor (optional)
        sensor_tangki_atas = not sensor_tangki_atas
        sensor_tangki_bawah = not sensor_tangki_bawah

except KeyboardInterrupt:
    print("\nProgram dihentikan.")
```
